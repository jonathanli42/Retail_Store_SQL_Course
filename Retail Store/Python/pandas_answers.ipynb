```python
import pandas as pd
```

```python
clicks_df = pd.read_csv('../Data/clicks_data.csv')
customers_df = pd.read_csv('../Data/customers_data.csv')
discount_df = pd.read_csv('../Data/discount_data.csv')
marketing_df = pd.read_csv('../Data/marketing_data.csv')
orders_df = pd.read_csv('../Data/orders_data.csv')
product_df = pd.read_csv('../Data/product_data.csv')
returns_df = pd.read_csv('../Data/returns_data.csv')
```

```python
orders_df.head(5)
```

```python
#First and Last Purchase Date for Each Customer

orders_df.groupby('customer_id')['sales_date'].agg(
    first_purchase='min',
    last_purchase='max'
).reset_index()
```

```python
#Select All Products in Our Catalog
product_df['product_name'].unique()
```

```python
#Find Specific Products with a Sales Price Greater Than $50

product_df[product_df['sales_price'] > 50]
```

```python
#Count of Orders for Each Customer
orders_df.groupby('customer_id')['order_id'].agg(total_orders='nunique').reset_index()
```

```python
#Total Number of Unique Customers That Have Purchased from Us

orders_df['customer_id'].nunique()
```

```python
#Total Number of Unique Customers by Product

orders_df.groupby('product_id')['customer_id'].agg(total_customers='nunique').reset_index()
```

```python
#Summarize Sales by Product

product_sales = orders_df.groupby('product_id')['total_sales'].agg(total_sales='sum').reset_index()

product_sales.merge(product_df, on='product_id', how='left')
```

```python
#Summarize Sales by Month
orders_df_copy = orders_df.copy()
orders_df_copy['sales_date'] = pd.to_datetime(orders_df_copy['sales_date'])


orders_df_copy.groupby(orders_df_copy['sales_date'].dt.to_period('M'))['total_sales'].agg(total_sales='sum').reset_index()
```

```python
#Find Orders Within a Date Range - 2023-01-01 - 2023-06-01

columns = ['order_id', 'sales_price', 'units', 'total_sales', 'customer_id']
orders_df_copy[(orders_df_copy['sales_date'] >= '2023-01-01') & (orders_df_copy['sales_date'] <= '2023-06-01')][columns].reset_index()
```

```python
#Customers with No Orders

"""
Get customers with orders > 1 
"""
    
customer_orders = pd.DataFrame(orders_df['customer_id'].unique())

customers_df[~customers_df['customer_id'].isin(customer_orders)]
```

```python
#Total Sales Per Customer

orders_df.groupby('customer_id')['total_sales'].agg(total_sales_by_cust='sum').reset_index()
```

```python
#Average Sales Price of Products in Each Category

orders_df.groupby('product_id')['total_sales'].agg(avg_sales='mean').reset_index()
```

```python
#Find the Top 5 Products with the Highest Total Sales

orders_df.groupby('product_id')['total_sales'].agg(total_sum_product='sum').reset_index().head(5)
```

```python
#Find the Total Number of Returns Made by Each Customer

returns_df.groupby('customer_id')['return_id'].agg(total_returns='nunique').reset_index().sort_values(by='total_returns', ascending=False)
```

```python
#Total Clicks Grouped by Month

clicks_df_copy = clicks_df.copy()

clicks_df_copy['click_date'] = pd.to_datetime(clicks_df_copy['click_date'])


clicks_df_copy.groupby(clicks_df_copy['click_date'].dt.to_period('M'))['click_id'].agg(total_clicks='nunique').reset_index()
```

```python
#Total Clicks by Device Type

clicks_df.groupby('device_type')['click_id'].agg(total_clicks='nunique').reset_index()
```

```python
#Total Clicks from Campaigns
clicks_df.groupby('campaign_id')['click_id'].agg(total_clicks='nunique').reset_index()
```

```python
#Conversion Rate by Device Type (Clicks and Orders)
total_clicks_df = clicks_df.groupby('device_type')['click_id'].agg(total_clicks='nunique').reset_index()
total_orders_df = clicks_df.groupby('device_type')['order_id'].agg(total_orders='nunique').reset_index()
conversion_df = total_clicks_df.merge(total_orders_df, on='device_type', how='inner')
conversion_df['conversion_rate'] = (conversion_df['total_orders']/conversion_df['total_clicks'] * 100).round(2)
conversion_df['conversion_rate'] = conversion_df['conversion_rate'].astype(str) + '%'

conversion_df
```

```python
#Highest Total Sale for an Order & Total Orders at Highest Price
max_order_df = orders_df.groupby('order_id')['total_sales'].agg(max_sale='max').reset_index().sort_values(by='max_sale', ascending=False)
highest_sale = max_order_df['max_sale'].iloc[0]
max_order_df = max_order_df[max_order_df['max_sale'] == highest_sale]

max_order_df['order_id'].nunique()
```

```python
#Customers Who Have Ordered Every Product
total_products = product_df['product_id'].nunique()
cust_unique_customers = orders_df.groupby('customer_id')['product_id'].agg(total_products='nunique').reset_index()
cust_unique_customers[cust_unique_customers['total_products'] == total_products]
```
